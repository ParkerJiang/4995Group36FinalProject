```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
clothing_df = pd.read_csv('articles.csv')
```

```python
clothing_df
```

```python
product_type_counts_df = clothing_df['product_type_name'].value_counts(ascending=False)
product_type_counts_df
ax = sns.barplot(x=product_type_counts_df.index[0:10], y=product_type_counts_df[0:10])
print("There are", product_type_counts_df.shape, "Product Types in total")
ax.set_xlabel("Product Type", fontsize = 12)
ax.set_ylabel("Count", fontsize = 12)
ax.set_title("Top Ten Product Types", fontsize = 14)
sns.set(rc = {'figure.figsize':(15,8)})
```

```python
color_group_counts_df = clothing_df['colour_group_name'].value_counts(ascending=False)
print("There are", color_group_counts_df.shape, "Color Groups in total")
ax = sns.barplot(x=color_group_counts_df.index[0:10], y=color_group_counts_df[0:10])
ax.set_xlabel("Color Group", fontsize = 12)
ax.set_ylabel("Count", fontsize = 12)
ax.set_title("Top Ten Color Groups", fontsize = 14)
```

```python
clothing_types_df = clothing_df['index_name'].value_counts(ascending=False)
print("There are", clothing_types_df.shape, "Clothing Types in total")
ax = sns.barplot(x=clothing_types_df.index[0:10], y=clothing_types_df[0:10])
ax.set_xlabel("Clothing Type", fontsize = 12)
ax.set_ylabel("Count", fontsize = 12)
ax.set_title("Top Ten Clothing Types", fontsize = 14)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 10, rotation=45)
print()
```

```python
index_group_name_df = clothing_df['index_group_name'].value_counts(ascending=False)
print("There are", index_group_name_df.shape, "Broad Clothing Types in total")
ax = sns.barplot(x=index_group_name_df.index[0:10], y=index_group_name_df[0:10])
ax.set_xlabel("Broad Clothing Type", fontsize = 12)
ax.set_ylabel("Count", fontsize = 12)
ax.set_title("Top Broad Clothing Types", fontsize = 14)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 10, rotation=45)

print("There is a much larger amount of ladieswear/childrenswear than there is menswear")
```

```python
customers_df = pd.read_csv('customers.csv')
```

```python
customers_df
```

```python
ax = sns.boxplot(x=customers_df["age"])
ax.set_title("Distribution of Customer Age", fontsize = 14)
```

```python
club_member_status_df = customers_df['club_member_status'].value_counts(ascending=False)
print(club_member_status_df.head())
print("There are", club_member_status_df.shape, "Club Member Status types in total")
ax = sns.barplot(x=club_member_status_df.index, y=club_member_status_df)
ax.set_xlabel("Club Member Status", fontsize = 12)
ax.set_ylabel("Count", fontsize = 12)
ax.set_title("Club Member Status", fontsize = 14)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 10)

print("Most customers in this dataset are active members in the rewards program")
```

```python
fashion_news_frequency_df = customers_df['fashion_news_frequency'].value_counts(ascending=False)
print(fashion_news_frequency_df.head())
print("There are", fashion_news_frequency_df.shape, "Fashion News Frequency Categories in total")
ax = sns.barplot(x=fashion_news_frequency_df.index, y=fashion_news_frequency_df)
ax.set_xlabel("Fashion News Frequency", fontsize = 12)
ax.set_ylabel("Count", fontsize = 12)
ax.set_title("Fashion News Frequency", fontsize = 14)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 10)
print("Most customers in this dataset do not subscribe to fashion news")

#Two none columns, maybe combine them.
```
